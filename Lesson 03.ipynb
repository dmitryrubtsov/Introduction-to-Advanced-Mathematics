```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def polar(R, a):
    x = R * np.cos(a)
    y = R * np.sin(a)
    return x, y
```

```python
t = np.arange(-2 * np.pi, 2 * np.pi, 0.01)
r = 1
x = r * np.cos(t)
y = r * np.sin(t)

plt.polar(x, y)
plt.show()
```

1. Решите систему уравнений: 
$$ \left\{
\begin{array}{l}
y = x^2 – 1 \\
exp(x) + x∙(1 – y) = 1
\end{array}
\right .$$  

1. Решите систему уравнений и неравенств:
$$\left \{
\begin{array}{l}
y = x^2 – 1 \\
exp(x) + x∙(1 – y) > 1
\end{array}
\right .$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-6, 6, 200)
y = x**2 - 1

plt.plot(x, y, label=f'$y = x^2 – 1$')

y = (((np.exp(x)) - 1) / x) + 1
plt.plot(x, y, label=f'$exp(x) + x∙(1 – y) = 1$')

plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.linspace(-3, 5, 200)

y = np.exp(x) - 1 + x*(2-x**2)
plt.plot(x, y)
plt.grid(True)
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from scipy.optimize import fsolve
```

```python
def equations(x):
    return np.exp(x) - 1 + x*(2-x**2)
```
